```python
#importing librarires

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

```

```python
#pip install scikit-learn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Jeevita-R/ML_datasets/refs/heads/main/Bengaluru_House_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()# col wise sum- of null
df=df.drop(["area_type","availability","society","balcony"],axis=1)
```

```python
df.head()
```

```python
df['location'].value_counts()#
df['location']=df['location'].fillna('Whitefield')
df.info()# handled location null value 
```

```python
df['size'].value_counts()# 2 bhk repeated for more times we fill it in all missing value
df['size']=df['size'].fillna('2 BHK')
df.info()
```

```python
# handling missing values in bath column
df['bath']=df['bath'].fillna(df['bath'].median())
df.info()
```

```python
df.head()
```

```python
# cleaning size column
df['size'].unique()
out=[i.split()[0] for i in df['size']]
df['bhk']=out
df.head()
```

```python
df.info()
# convert bhk col from obj to int
df['bhk']=df['bhk'].astype('int')
df.info()
```

```python
df.head()
```

```python
# cleaning total sq feet

df['total_sqft'].unique()
def convert(a):
    l=a.split('-')
    if len(l)==2:
        return (float(l[0]) + float(l[1])) / 2
    try:
        return float(a)
    except:
        return None
df['total_sqft'] = df['total_sqft'].apply(convert)
df['total_sqft']=df['total_sqft'].fillna(df['total_sqft'].median())
df.info()  
        
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
df['total_sqft'].unique()
df['price_per_sqft']=df['price']*100000/df['total_sqft']
df['price_per_sqft']
df.head()
```

```python
df['location']=df['location'].apply(lambda x:x.strip())
df.head()
df['location'].value_counts()# deleting all trailing white spaces
```

```python
df['location'].value_counts()
loc_cnt = df['location'].value_counts()

# 2. Identify locations that appear 10 times or fewer
rare_locations = loc_cnt[loc_cnt <= 10].index

# 3. Replace rare locations with 'Other'
df['location'] = df['location'].replace(rare_locations, 'Other')

# 4. Check the updated counts
df['location'].value_counts()
```

```python
# further cleaning location
# identifying location whose count ir no of times they are apperaing are 10 or less than 10
# check previous cell done that
```

```python
df = df[(df['total_sqft'] / df['bhk']) >= 300]
df.describe()
```

```python
# handling outliers in total_sqft
df=df[df['total_sqft']>=df['bhk']*300]
df.describe()
```

```python
#handling outliers in price_per_sqft
sns.boxplot(y='price_per_sqft',data=df)
```

```python
#handling outliers in price_per_sqr
q1=df['price_per_sqft'].quantile(0.25)
q3=df['price_per_sqft'].quantile(0.75)
IQR=q3-q1
low_boun=q1-0.5*IQR
up_boun=q3+0.5*IQR
df=df[(df['price_per_sqft']>=low_boun)&(df['price_per_sqft']<=up_boun)]
sns.boxplot(y='price_per_sqft',data=df)
```

```python
df.head()
```

```python
df=df.drop(['size','price_per_sqft'],axis=1)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
df.head()# then i dropped the index
```

```python
df.head()
```

```python
copied_dataset=df.copy()
copied_dataset
```

```python
from sklearn.preprocessing import LabelEncoder
obj=LabelEncoder()
df['location']=obj.fit_transform(df['location'])
df.head()
```

```python
copied_dataset.head()
```

```python
y=df['price']
x=df.drop('price',axis=1)
x.drop('index',axis=1)
```

```python
x.drop('index',axis=1,inplace=True)
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestRegressor(random_state=42)
params={
"n_estimators":[100,150,200,250],
"max_depth":[3,4,5,6]    
}
```

```python
Grid=GridSearchCV(estimator=model,param_grid=params,cv=4)
Grid.fit(xtrain,ytrain)
```

```python
print('training efficiency:',Grid.score(xtrain,ytrain))
print('testing efficiency:',Grid.score(xtest,ytest))
```

```python
with open('RFmod.pkl','wb') as file:
    pickle.dump(Grid,file)
```
